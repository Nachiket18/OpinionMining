### Algorithm for Merging entities / aspects (compound)

1. If the relation with first noun and another word is ' nsubj ' :
   
*    If the recepient of the relation is Plural (eg 'are','were') then --> two different entitites / aspects . Example India and US **are** natural allies

2. If receipent of the relation is Noun (consideration for entity/aspect) and relation is 'poss' then we can conclude that receipient noun with conjuction with another noun is singleton. For e.g (China***'s*** **Belt and Road** *bold text*initiative is good.)

3. Multiple Compound words. These should be merged to get a single entity / aspect. for e.g Prime Minister Narendra Modi. Narendra --> Modi is compound, Minister --> Modi is Compound so Minister Narendra Modi is combined. Prime --> Minister Compound so we append Prime as well.  







Compound Word Extraction and Conjunction Identification Implementation

```python
def extract_compound_words(node):  
    subtree = node.head.subtree
    merged_word = None
    tmp_merge_word = None

    for nodes in subtree:
      if nodes.lemma_ == node.lemma_ and nodes.dep_ == 'compound':

          merged_word = nodes.lemma_ + ' ' + nodes.head.lemma_
          tmp_merge_word = nodes.lemma_
          tmp_merge_word_parent = nodes.head.lemma_

      elif tmp_merge_word != None and tmp_merge_word_parent == nodes.head.lemma_ :
          merged_word = tmp_merge_word + ' ' + nodes.lemma_ + ' ' + tmp_merge_word_parent

    if merged_word != None:
       nested_merge = extract_compound_words(node.head)
       if nested_merge != None:
         merged_word = node.lemma_ + ' ' + nested_merge

    
    return merged_word

  

def identify_conjunction_words(node):

    ## Point 1 of merging entity / aspect
    
    isConjunction = False

    subtree = node.subtree


    for subtree_node in subtree:
      if subtree_node.tag_ in target_tag and subtree_node.head.lemma_ == node.lemma_ and subtree_node.dep_ == 'conj':
        nested_subtree = subtree_node.head.subtree

        for nested_subtree_node in nested_subtree:
          if nested_subtree_node.lemma_ == node.lemma_ and nested_subtree_node.dep_ == 'nsubj' and nested_subtree_node.head.tag_  in ('VBP'):
            isConjunction = False
          
          elif nested_subtree_node.lemma_ == node.lemma_ and nested_subtree_node.dep_ == 'nsubj' and nested_subtree_node.head.tag_ not in ('VBP'):
            if nested_subtree_node.head.tag_ in ('VBD') and nested_subtree_node.head.text == 'was' :
              #print("--")
              #print("nest_node= ",nested_subtree_node.lemma_,nested_subtree_node.text,nested_subtree_node.tag_,nested_subtree_node.dep_,node.lemma_)
              #print(nested_subtree_node.head.text,nested_subtree_node.head,nested_subtree_node.head.tag_)
              isConjunction = True
              return isConjunction

    ## Point 2 
    
      if subtree_node.tag_ in target_tag and subtree_node.lemma_ == node.lemma_ and subtree_node.dep_ == 'poss':
        #print(subtree_node.lemma_,subtree_node.text,subtree_node.tag_,subtree_node.dep_,subtree_node.head.text,subtree_node.head.tag_)
        isConjunction = True

      elif subtree_node.tag_ in ('PRP$','PRP') and subtree_node.head.lemma_ == node.lemma_ and subtree_node.dep_ == 'poss':
        #print('Captured')
        isConjunction = True


    ## Point 3

    return isConjunction
      


#doc = nlp("India and US was natural allies in terms of defense and economy")
#doc = nlp("China's Belt and Road Iniative was a great success.")
doc = nlp("Belt and Road Initiative is good")
#doc = nlp ("His Belt and Road initiative is really good")

for node in doc: 
  if node.tag_ in target_tag:
    res = identify_conjunction_words(node)
    print (node.lemma_,res)


#doc = nlp(" Prime Minister Narendra Modi is strong leader ")

#doc = nlp("Belt and Road Initiative is good")
#for node in doc:
#  if node.lemma_ in ('Road'): 
#    merged_word = identify_conjunction_words(node)
#    print(merged_word)


```
